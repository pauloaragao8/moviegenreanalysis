**Drama - Introduction**

According to the MasterClass education, the drama genre features stories with 
high stakes and a lot of conflicts. 

They’re plot-driven and demand that every character and scene move the story forward. 
Dramas follow a clearly defined narrative plot structure, portraying real-life 
scenarios or extreme situations with emotionally-driven characters. Films that 
fall into drama sub-genres include historical drama, romantic drama, teen drama, 
medical drama, docudrama, and film noir.

Also, according to MasterClass staff, Drama is a form of writing that originated in ancient Greece. 
While today’s dramas have moved from Greek amphitheaters to Hollywood sound stages, the 
basic elements of writing a drama screenplay still remain the same.

Furthermore, the website FilmSite explains that Drama marks the division between tragedy and comedy. 
A drama can take place anywhere with any characters in any time period; it’s not the setting that 
makes a movie a drama, but the tone. Dramas are serious, full of conflict, 
and continually push the boundaries of the plot by raising the stakes. 
Depending on the sub-genre of the drama, the story can range from a forbidden 
romance to a murder investigation.

**Themes of Drama**

   Dramatic themes often include current issues, societal ills, and problems, 
concerns or injustices, such as racial prejudice, religious intolerance 
(such as anti-Semitism), drug addiction, poverty, political unrest, the 
corruption of power, alcoholism, class divisions, sexual inequality, mental 
illness, corrupt societal institutions, violence toward women or other 
explosive issues of the times, as cited by FilmSite.

   According to our dataset, we could identify different movie genres and other aspects.

### Analysis on Correlation of Variables in Drama

As seen in the correlation graph above, the average Rating of Drama is more tighly connected to Runtime than it

While the correlation between number of votes and ratings or runtime is weaker (0.04 and 0.07 respectively).

```python
# instantiating an lmplot for year and runtime of Drama Movies
sns.lmplot(x          = 'year',  #defining the x axis
           y          = 'runtimeMinutes',  #defining the y axis
           hue        = 'runtimeMinutes', # formats legend if hue != None
           legend     = False,        # supressing the legend
           legend_out = False,    #excluding legend
           scatter    = True,     #setting scatter type
           fit_reg    = True,     #estimating a regression model
           aspect     = 2,        #defining aspect ratio
           data       = drama_movies) #defining dataset


# developing vertical axis lines
value = 10

while value < 2021:
    
    # making a vertical line
    plt.axvline(x = value, color = "purple", linestyle = '--')
    
    # incrementing value in one-quarter increments
    value += 10


# formatting and displaying the plot
plt.title(label    = 'Year and Runtime')
plt.xlabel(xlabel  = 'Runtime')
plt.ylabel(ylabel  = 'Runtime')
plt.xlim(1950, 2021)
plt.tight_layout()
plt.savefig('drama_year.png')
plt.show()
```
